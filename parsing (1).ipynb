```python
def transfrom_metrices(name, unit, added_metrices, pre_string):
    name = name.replace(' ', '_')
    if unit == 'Avg':
        name+='_avg'
    elif unit == 'Min':
        name += '_min'
    elif unit == 'Max':
        name += '_max'
    elif unit == 'Mean':
        name += '_mean'
    elif unit == 'Std Dev':
        name += '_std'

    
    name = pre_string + '_' + name
    return name
```
