```python
df = pd.DataFrame()
df_2 = pd.DataFrame()
for k in [0]:
    for run in range(0,7):
        if run==4:
            continue
        temp_df = pd.read_csv('amd_l2_cache_' + str(run) + '_k' + str(k))
        if run == 0:
            df = temp_df
            df['Value'] = df['Value'].astype(str) 
            continue
        for index, row in temp_df.iterrows():
            # print(row['Metrics'])
            # print(df.loc[df['Metrics']==row['Metrics'],'Value'])
            df.loc[df['Metrics']==row['Metrics'],'Value'] += ',' + str(row['Value'])
    #     break
    # break
for k in [1]:
    for run in range(0,7):
        if run==4:
            continue
        temp_df = pd.read_csv('amd_l2_cache_' + str(run) + '_k' + str(k))
        if run == 0:
            df_2 = temp_df
            df_2['Value'] = df_2['Value'].astype(str) 
            continue
        for index, row in temp_df.iterrows():
            df_2.loc[df_2['Metrics']==row['Metrics'],'Value'] += ',' + str(row['Value'])
    #     break

df_2 = df_2.rename(columns={'Value': 'Value_2'})
# df_2 = df_2.set_index('Metrics')
# print(df_2.head())
# df = df.set_index('Metrics')
# result = pd.concat([df_2, df], axis=1, join="inner")
result = df_2.merge(df,on='Metrics')
result.set_index('Metrics')
result = result.drop(['Unnamed: 0_x','Unnamed: 0_y'],axis=1)

df = pd.read_csv('combinedworkloads_runtime.csv')

for index, row in result.iterrows():
    elements1 = [float(x) for x in row['Value_2'].split(',')]
    res1 = all(x == 0 for x in elements1)
    elements2 = [float(x) for x in row['Value'].split(',')]
    res2 = all(x == 0 for x in elements2)
    if res1 and res2:
        result = result.drop(index)
    print(row)

# print(result)
# result.to_csv('test.csv')
```
